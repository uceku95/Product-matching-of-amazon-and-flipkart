**Installing thefuzz**

```python
from thefuzz import fuzz,process
```

```python
amazonpn=print("amazon proudctname:", process.extract("Alisha Solid Women's Cycling Shorts", df1.amazon_product_name, limit=1, scorer=fuzz.token_sort_ratio) )
amazonrp=print("amazon retail_price:", process.extract("Alisha Solid Women's Cycling Shorts", df1.amazon_retail_price, limit=1, scorer=fuzz.token_sort_ratio) )
amazondp=print("amazon discounted_price:", process.extract("Alisha Solid Women's Cycling Shorts", df1.amazon_discounted_price, limit=1, scorer=fuzz.token_sort_ratio) )
flipkartpn=print("Flipkart_product_name:", process.extract("Alisha Solid Women's Cycling Shorts", df2.Flipkart_product_name, limit=1, scorer=fuzz.token_sort_ratio) )
flipkartrp=print("Flipkart_retail_price:", process.extract("Alisha Solid Women's Cycling Shorts", df2.Flipkart_retail_price, limit=1, scorer=fuzz.token_sort_ratio) )
flipkartdp=print("Flipkart_discounted_price", process.extract("Alisha Solid Women's Cycling Shorts", df2.Flipkart_discounted_price, limit=1, scorer=fuzz.token_sort_ratio) )
```
